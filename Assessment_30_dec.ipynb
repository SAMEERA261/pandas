Create  a class ‘Car261_264’ which takes a url/path as an input to create an object. Assert if the columns in the data frame are same as [  
* id_1
* id_2
* route
* moto
* car
* rv
* bus
* truck

] and in the same order. This class should have five methods (names given below) which performs the said functionality.


```python
import pandas as pd,numpy as np

class Car261_264:
    def __init__(self, url):
        self.df = pd.read_csv(r"C:\Users\Sam\Downloads\dataset-1.csv")
        self.assert_columns()

    def assert_columns(self):
        expected_columns = ["id_1", "id_2", "route", "moto", "car", "rv", "bus", "truck"]
        assert list(self.df.columns) == expected_columns, "Columns do not match."

    def generate_car_matrix(self):
        
        columns = self.df['id_2'].unique()
        index = self.df['id_1'].unique()
        car_matrix = pd.DataFrame(index=index, columns=columns)
        for _, row in self.df.iterrows():
            car_matrix.at[row['id_1'], row['id_2']] = row['car']
        return car_matrix
    
    def get_type_count(self):
        self.assert_columns() 
        self.df['car_type'] = pd.cut(self.df['car'], bins=[-float('inf'), 15, 25, float('inf')],
                                     labels=['low', 'medium', 'high'], include_lowest=True, right=False)
        type_count = self.df['car_type'].value_counts().to_dict()
        type_count = dict(sorted(type_count.items()))
        return type_count

    def get_bus_indexes(self):
        self.assert_columns()
        mean_bus = 2 * self.df['bus'].mean()
        bus_indexes = self.df[self.df['bus'] > mean_bus].index.tolist()
        return sorted(bus_indexes)

 
    def filter_routes(self):
        self.assert_columns()
        filtered_routes = self.df.groupby('route')['truck'].mean()
        filtered_routes = filtered_routes[filtered_routes > 7].index.tolist()
        return sorted(filtered_routes)


    def multiply_matrix(self, car_matrix):
        modified_matrix = car_matrix.applymap(lambda x: x * 0.75 if x > 20 else x * 1.25).round(1)
        return modified_matrix
```
