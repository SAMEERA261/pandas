#### The main diffrence between cheking accounts and saving accounts is that -
* checking account is being used for all type of transactions and can be used  for deposits,pay bills,transfer money between accounts and they have no limits on online transfers ,ATM transactions,cheques,and debit card.
* savings accounts have no online bill payments no debit cards payments:withdrawal and online transfers are limited to only 6 times a month.its not subjected to risk savings will be done in small amounts,and will be earned small amount of intrest .
### in the above data we can see the below columns that have been taken for analysis:
* Age (numeric)
* Sex (text: male, female)
* Job (numeric: 0 - unskilled and non-resident, 1 - unskilled and resident, 2 - skilled, 3 - highly skilled)
* Housing (text: own, rent, or free)
* Saving accounts (text - little, moderate, quite rich, rich)
* Checking account (numeric, in DM - Deutsch Mark)
* Credit amount (numeric, in DM)
* Duration (numeric, in month)
* Purpose (text: car, furniture/equipment, radio/TV, domestic appliances, repairs, education, business, vacation/others)

Managing and mitigating credit risk is crucial for lenders to avoid financial losses and ensure the stability of their operations.
