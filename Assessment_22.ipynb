### Step 8. Present the mean preTestScores grouped by regiment and company without hierarchical indexing.

```python
mean_pretest_scores = rd.groupby(['regiment', 'company'])['preTestScore'].mean().unstack()#to convert row index to a column lable
```

```python
mean_pretest_scores
```

* The unstack() function in pandas is used to pivot a level of the row (index) labels to the columns. It is often used with DataFrames that have a multi-level index resulting from a groupby operation.
* The unstack() function is then applied to move one level of the multi-level index to become a new level of column labels. This is useful for reshaping the DataFrame.

### Step 9. Group the entire dataframe by regiment and company.

```python
a = regiment.groupby(['regiment', 'company']).apply(lambda x: x.reset_index(drop=True))
a
```

* lambda x: x.reset_index(drop=True): The lambda function takes each group (x), and the reset_index(drop=True) part is used to reset the index of each group. The drop=True parameter is used to drop the existing index column and replace it with a new one.
* reset_index: This method resets the index of a DataFrame. The drop=True parameter is used to remove the old index column.
